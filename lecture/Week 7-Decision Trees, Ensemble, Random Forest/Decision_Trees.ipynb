## Load data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Dataset/adult.csv') #load the csv dataset into a 'dataframe' df

df.head()
```

```python
df[' <=50K'].unique() #there are only two classes
```

## Data preprocessing
* Rename the columns

```python
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship',
             'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']

df.columns = col_names

df.head()
```

## Prepare the train and test data

```python
from sklearn.model_selection import train_test_split

```

```python

categorical_columns =['workclass', 'education', 'marital_status', 'occupation', 'relationship',
                                 'race', 'sex', 'native_country']

#One hot encoding for categorical attributues/features
df_encode = pd.get_dummies(data = df, prefix = 'OHE', prefix_sep='_',
               columns = categorical_columns,
               drop_first =True,
              dtype='int8')

df_encode.head()


# Prepare the training and testing dataset
X = df_encode.drop('income',axis=1) # Independant variable
y = df_encode['income'] # dependant variable


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
cols = X_train.columns


#Scaling the dataset.
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


X_train = pd.DataFrame(X_train, columns=[cols])
X_test = pd.DataFrame(X_test, columns=[cols])


X_train.head()
```

## Train the model

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# Instantiate the model
clf = DecisionTreeClassifier(random_state=0, criterion='gini')

# Fit the model
clf.fit(X_train,y_train)

# Make predictions with the model
y_pred = clf.predict(X_test)

# Evaluate the trained model
from sklearn.metrics import accuracy_score
print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
